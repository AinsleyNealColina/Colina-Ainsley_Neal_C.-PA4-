```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame = pd.read_excel("C:\\Users\\DELL\\Downloads\\board2.xlsx")
df = pd.DataFrame(dataFrame, columns=['Name', 'Gender', 'Track', 'Hometown', 'Math','Electronics','GEAS'])
display (df)
```

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS']].mean(axis=1)

# Set the style for the plots
sns.set(style="whitegrid")

# Bar plot for average grades by track
plt.figure(figsize=(10, 6))
sns.barplot(x='Track', y='Average', data=df, ci=None)
plt.title('Average Grade by Track')
plt.show()

# Bar plot for average grades by gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Average', data=df, ci=None)
plt.title('Average Grade by Gender')
plt.show()

# Bar plot for average grades by hometown
plt.figure(figsize=(10, 6))
sns.barplot(x='Hometown', y='Average', data=df, ci=None)
plt.title('Average Grade by Hometown')
plt.show()

# Box plot for average grades by track
plt.figure(figsize=(10, 6))
sns.boxplot(x='Track', y='Average', data=df)
plt.title('Distribution of Average Grades by Track')
plt.show()

# Box plot for average grades by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Average', data=df)
plt.title('Distribution of Average Grades by Gender')
plt.show()

# Box plot for average grades by hometown
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hometown', y='Average', data=df)
plt.title('Distribution of Average Grades by Hometown')
plt.show()
```
